finding Links form webPage

```python
import requests
from bs4 import BeautifulSoup
 
 
url = 'https://en.wikipedia.org/wiki/Main_Page'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.content, 'html.parser')

```

```python
print(soup.prettify())
```

```python
soup.a
```

```python
links = soup.find_all('a')

print(len(links))
```

```python
for link in links:
    if "http" in str(link):
        print(link)
```
