```python
import newspaper

newspaper_repository = newspaper.build('https://www.cnn.com/', memoize_articles=False, language='en')
print(len(newspaper_repository.articles))

for article in newspaper_repository.articles:
    print(article.url)
    
for category in newspaper_repository.category_urls():
    print(category)
    
```

```python
import json
data = {}
data['article'] = []
for articles in newspaper_repository.articles:
    retrive_article = Article(article.url)
    try:
       retrive_article.download() 
       retrive_article.parse()
       if retrive_article.text:
        data['article'].append({
            "title":retrive_article.title,
            "url":retrive_article.url,
            "date":retrive_article.publish_date,
            "text":retrive_article.text
        })
    except:
        print(article.url)
    
with open('NewsData.json', 'w') as f:
    json.dump(data, f)
```
